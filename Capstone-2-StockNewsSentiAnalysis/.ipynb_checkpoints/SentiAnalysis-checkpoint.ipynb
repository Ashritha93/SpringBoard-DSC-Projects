```python
import nltk
from nltk.tokenize import TweetTokenizer
from nltk.tokenize import word_tokenize
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
from textblob import TextBlob
```

```python
data = pd.read_csv('TGB_newsData.csv')
```

```python
data.head(8)
```

```python
df = pd.DataFrame(columns=['HeadlinesBag', 'NewsBag'])
df
```

```python
data['News'] = data['News'].fillna(data['Headlines'])
data.head(8)
```

```python
# data['Headlines'] = data['Headlines'].str.lower()
# data['News'] = data['News'].str.lower()
```

```python
df['HeadlinesBag'] = data['Headlines'].apply(lambda k : word_tokenize(k))
df['NewsBag'] = data['News'].apply(lambda k : word_tokenize(k))
df
```

```python
def get_textBlob_score(sent):
    polarity = TextBlob(sent).sentiment.polarity
    return polarity
```

```python
data['Headlines_Score_tb'] = [get_textBlob_score(i) for i in data['Headlines']]
data['Headlines_Score_tb']
```

```python
data['News_Score_tb'] = [get_textBlob_score(i) for i in data['News']]
data['News_Score_tb']
```

```python
sid = SentimentIntensityAnalyzer()
def get_vader_score(sent):
    ss = sid.polarity_scores(sent)
    #for k in sorted(ss):
        #print('{0}: {1}, '.format(k, ss[k]), end='')
        #print()
    return ss['compound']
```

```python
data['Headlines_vader'] = [get_vader_score(i) for i in data['Headlines']]
data['News_vader'] = [get_vader_score(i) for i in data['News']]
```

```python
data
```

```python
def get_Vsentiment(x):
    if x >= 0.05:
        return int(1)
    elif x > -0.05 and x < 0.05:
        return int(0)
    else :
        return int(-1)

```

```python
# Polarity scores of vader
# 1 if i >= 0.05
# 0 if i > -0.05 and i < 0.05
# -1 if i <= -0.05
data['Head_vader_senti'] = data['Headlines_vader'].apply(lambda x: get_Vsentiment(x))
data[['Headlines_vader', 'Head_vader_senti']]
```

```python
data['News_vader_senti'] = data['News_vader'].apply(lambda x: get_Vsentiment(x))
data[['News_vader', 'News_vader_senti']]
```

```python
# Decoding textblob sentiment analysis:
def getTextBS(x):
    if x > 0 :
        return int(1)
    elif x == 0 :
        return int(0)
    else:
        return(-1)
```

```python
data['Headlines_textB_s'] = data['Headlines_Score_tb'].apply(lambda x: getTextBS(x))
data['News_textB_s'] = data['News_Score_tb'].apply(lambda x: getTextBS(x))
```

```python
data
```

```python
df = pd.DataFrame(data[['Head_vader_senti', 'News_vader_senti', 'Headlines_textB_s', 'News_textB_s']])
```
