```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = ['https://www.moneycontrol.com/stocks/company_info/stock_news.php?sc_id=TT&scat=&pageno=1&next=0&durationType=M&Year=&duration=6&news_type='
       , 'https://www.moneycontrol.com/stocks/company_info/stock_news.php?sc_id=TT&scat=&pageno=2&next=0&durationType=M&Year=&duration=6&news_type=']
```

```python
html_content = [requests.get(i).text for i in url]
```

```python
soup = dict()
for i in range(len(html_content)):
    soup[i] = BeautifulSoup(html_content[i], "lxml")
# print(soup[0].prettify())
soup.keys()
```

```python
for i in soup.keys():
    print (i)
```

```python
data = {}
# Firstly, we need to extract all the available headlines of news from the page
get_news = [soup[i].find_all('div', attrs={'class':'MT15 PT10 PB10'}) for i in soup.keys()]
get_news_headline = [j.select('strong') for i in get_news for j in i]
get_news_headlines_final = [i[0].get_text() for i in get_news_headline if len(i) > 0]

get_news_headlines_final
```

```python
# Secondly data of the news
get_news_time = [j.select('p', attrs={'class':'PT3 a_10dgry'}) for i in get_news for j in i]
get_news_time_final = [i[0].get_text().split('|') for i in get_news_time if len(i) > 0]
get_news_time_final
# [0:22]
```

```python
data = pd.DataFrame.from_records(get_news_time_final, columns=['Time', 'Date', 'SourceOfNews'])
data['Headlines'] = get_news_headlines_final
data
```

```python
get_news_links = [j.select('a', attrs={'class':'g_14bl'}) for i in get_news for j in i]
get_news_links_final = []
for i in get_news_links:
    for j in i:
        get_news_links_final.append(j['href'])
# print(get_news_links_final)
# series_links = pd.Series(get_news_links_final)
series_links1 = []
for i in range(len(get_news_links_final)-1):
    if get_news_links_final[i] == get_news_links_final[i+1]:
        series_links1.append(get_news_links_final[i])
# series_links1 = pd.Series(series_links1)
series_links1 = ['https://www.moneycontrol.com' + i for i in series_links1]
(series_links1)
```

```python
article = dict()

for i in range(len(series_links1)):
    html_contents = requests.get(series_links1[i]).text
    soup = BeautifulSoup(html_contents, "lxml")
    print(series_links1[i])
    placeHolder = [a.find_all('p') for a in soup.find_all('div', attrs={'class':'arti-flow'})]
    placeHolder = [a.get_text() for b in placeHolder for a in b]
    print(placeHolder, '\n')
    article[series_links1[i]] = placeHolder

# article
```

```python
len(article.keys())
```

```python
df = pd.DataFrame(article.values(), article.keys())
df = df[df.columns[1:]].apply(lambda x: '. '.join(x.dropna().astype(str)), axis=1)
df = df.reset_index()
df.columns = ['URL', 'News']
df
```

```python
data['News'] = df['News']
data
```

```python
data.to_csv(r'TGB_newsData.csv')
```
